```python
import pandas as pd

pd.set_option("display.max_colwidth", 100)
```

```python
df = pd.read_csv("sample_text.csv")
df.shape
```

```python
df
```

```python
from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df.text)
vectors.shape
```

```python
dim = vectors.shape[1]
dim
```

```python
import faiss

index = faiss.IndexFlatL2(dim)
index
```

```python
index.add(vectors)
```

```python
search_query = "An apple a day keeps the doctor away"

vec = encoder.encode(search_query)
vec.shape
```

```python
vec
```

we need to convert above vector into 2 dimentions of array. bcoz, this search vector expects 2 dimentional array

```python
import numpy as np
# search vector
svec = np.array(vec).reshape(1, -1)
```

```python
svec
```

```python
distances, I = index.search(svec, k=2)
# k: how many vectors do you need (it's like K-Nearest Neighbor (KNN))
```
